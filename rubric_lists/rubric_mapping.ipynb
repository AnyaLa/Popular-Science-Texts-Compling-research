```python
def mapping(tags_list):
    rubrics_list = []
    for tags_text in tags_list:
        try:
            rubrics_text = []
            for tag in tags_text.split('_'):
                try:
                    rubric = tag_map[tag.lower().strip()]
                    rubrics_text.append(rubric)   
                except KeyError:
                    pass
            rubrics_list.append(sorted(rubrics_text))
        except:
            rubrics_list.append([])
    return rubrics_list
```

```python
from collections import Counter
```

```python
def rubrication(rubrics_list):
    main_rubrics = []
    for rubrics_text in rubrics_list:
        #r_count = Counter(rubrics_text.split('|'))
        rubr = [r for r in rubrics_text.split('|') if (r!= 'Мусор')&(r!= '')]
        if rubr == []:
            rubr.append('Мусор')
        r_count = Counter(rubr)
        r_dict = dict(r_count)
        max_value = max(r_dict.values())
        rubrics = [rubric for rubric, value in r_dict.items() if value == max_value]
        main_rubrics.append(('|').join(sorted(rubrics)))
    return main_rubrics       
```

```python
def easy_rubrication(rubrics_list):
    rubrication = []
    for rubr in rubrics_list:
        if '|' not in rubr:
            rubrication.append(rubr)
        else:
            if 'Футурология' in rubr:
                rubrication.append('Футурология')
                continue
            if 'Космос' in rubr:
                rubrication.append('Космос')
                continue
            if ('Технологии') in rubr:
                rubrication.append('Технологии')
                continue
            if ('Computer Science') in rubr:
                rubrication.append('Computer Science')
                continue
            if ('Биология' in rubr) and ('Физиология человека' in rubr):
                rubrication.append('Физиология человека')
                continue
            if ('Психология') in rubr:
                rubrication.append('Психология')
                continue
            if ('Физика') in rubr and ('Химия' in rubr or 'Физиология человека' in rubr):
                rubrication.append('Физика')
                continue
            rubrication.append(rubr)
            
    return rubrication
```
