```python
metadata = pd.read_csv('joint_metadata.tsv', encoding = 'utf-8', index_col = 0, sep = '\t')
```

```python
def mapping(tags_list):
    rubrics_list = []
    for tags_text in tags_list:
        try:
            rubrics_text = []
            for tag in tags_text.split('_'):
                try:
                    rubric = tag_map[tag.lower()]
                    rubrics_text.append(rubric)   
                except KeyError:
                    pass
            rubrics_list.append(rubrics_text)
        except:
            rubrics_list.append([])
    return rubrics_list
```

```python
final_rubrics = mapping(metadata.tags)
```

```python
map_rubrics = []
for item in final_rubrics: 
    try:
        map_rubrics.append('|'.join(item))
    except:
        map_rubrics.append('')
```

```python
metadata["mapped_rubrics"] = map_rubrics
```

```python
metadata.head(50)
```

```python
from collections import Counter
```

```python
def rubrication(rubrics_list):
    main_rubrics = []
    for rubrics_text in rubrics_list:
        #r_count = Counter(rubrics_text.split('|'))
        rubr = [r for r in rubrics_text.split('|') if r!= 'Мусор']
        if rubr == []:
            rubr.append('Мусор')
        r_count = Counter(rubr)
        r_dict = dict(r_count)
        max_value = max(r_dict.values())
        rubrics = [rubric for rubric, value in r_dict.items() if value == max_value]
        main_rubrics.append(('|').join(rubrics))
    return main_rubrics       
```

```python
metadata['final_rubrics'] = rubrication(metadata.mapped_rubrics)
```
