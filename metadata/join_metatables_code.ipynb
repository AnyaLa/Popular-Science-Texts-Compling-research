```python
raw_dates_series = joint_meta['date']
```

```python
def unify_dates(raw_dates):
    parsed_dates = []
    for raw_date in tqdm(raw_dates):
        parsed_date = str(date.parse(raw_date, languages=['ru']))
        parsed_date = parsed_date[:10]
        parsed_dates.append(parsed_date)
    dates_series = pd.Series(parsed_dates)
    return dates_series
```

```python
dates_series = unify_dates(raw_dates_series)
```

```python
joint_meta['date'] = dates_series
```
