```python
def unify_dates(raw_dates):
    parsed_dates = []
    for raw_date in tqdm(raw_dates):
        if '-' in raw_date:
            parsed_eng = str(date.parse(raw_date))
            parsed_eng = parsed_eng[:10]
            parsed_dates.append(parsed_eng)
        if raw_date == None:
            parsed_dates.append(raw_date)
        else:
            parsed_ru = str(date.parse(raw_date, languages=['ru']))
            parsed_ru = parsed_ru[:10]
            parsed_dates.append(parsed_ru)
    dates_series = np.array(parsed_dates)
    return dates_series
```

```python
dates_series = unify_dates(raw_dates_series)
```

```python
joint_meta['date'] = dates_series
```
