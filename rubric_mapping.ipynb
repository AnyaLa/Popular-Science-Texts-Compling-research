```python
metadata = pd.read_csv('D:/Анна/Desktop/joint_metadata.csv', encoding = 'utf-8', index_col = 0)
```

```python
metadata.head(100)
```

```python
def mapping(tags_list):
    rubrics_list = []
    for tags_text in tags_list:
        try:
            rubrics_text = []
            for tag in tags_text.split('_'):
                try:
                    rubric = tag_map[tag]
                    rubrics_text.append(rubric)   
                except KeyError:
                    pass
            rubrics_list.append(rubrics_text)
        except:
            rubrics_list.append([])
    return rubrics_list
```

```python
final_rubrics = mapping(metadata.tags)
```

```python
map_rubrics = []
for item in final_rubrics: 
    try:
        map_rubrics.append('_'.join(item))
    except:
        map_rubrics.append('')
```

```python
map_rubrics[20000:21500]
```

```python
metadata["mapped_rubrics"] = map_rubrics
```
